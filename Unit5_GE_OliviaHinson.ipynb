## Part 1 - Definitions

+ **Visualizations**: A visually pleasing representation (graphs, charts, etc.) of data to explain relationships between them
+ **Matplotlib**: Cross-platform for two-dimensional plots frrom data in arrays
+ **Seaborn**: Similar to Matplotlib, but can provide higher functionality for common statistical plot types and integrates Pandas DataFrames
+ **Pandas Built-in Visualizations**: Represent basic plots such as bar plots, histograms, box plots, area plots, and scatter plots.

## Part 2 - Using Matplotlib

```python
# Present sine and cosine wave on same chart
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# Determine x values to use 
x = np.arange(0, 4 * np.pi, 0.1)

fig = plt.figure()
plt.title('sine & cosine wave forms') # Print title of plot
plt.plot(x, np.sin(x), '--') # Print with dashes
plt.plot(x, np.cos(x), '.') # Print with dots
```

## Part 3 - Using Seaborn & Pandas Built-In Visulizations

```python
# Create Seaborn and Pandas Visualizations
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set() # Set Seaborn plot styles

# Load Dataset to use 
mpg = sns.load_dataset('mpg')

# Seaborn linear regressions using regplot() and lmplot()
print('First plot use regplot()\nSecond plot use lmplot()\nThird plot use Pandas visulation with scatterplot')
print('Fourth plot use lmplot() with discrete values')
print('--------------------------------------------------')
sns.regplot(x = 'weight', y = 'acceleration', data = mpg)
sns.lmplot(x = 'weight', y = 'acceleration', data = mpg)

# Create Pandas scatterplot 
mpg.plot.scatter(x = 'weight', y = 'acceleration')

# Create linear regression with discrete values
sns.lmplot(x = 'cylinders', y = 'mpg', data = mpg)
plt.show
```

```python
# Create Seaborn and Pandas Visualizations
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
sns.set() # Set Seaborn plot styles

# Load daatset to use
exercise = sns.load_dataset('exercise')

# Plot PairGrid using different colors for each dataset
g = sns.PairGrid(exercise, hue = 'kind')
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter);

# PLot FacetGrid for histogram with associated 'pulse'
g = sns.FacetGrid(exercise, col = 'kind')
g.map(plt.hist, 'pulse')
plt.show()

# Create Pandas histograms
exercise.plot.hist(stacked = True, bins = 20) # Stacked with 20 bins
exercise.plot.hist(stacked = False, bins = 20) # Not stacked with 20 bins
exercise.plot.hist(alpha = 0.5, bins = 20)
```
